#Very basic plot to visualize any data

scatter plot  

ใช้จุดแสดง dimension

```
pollution_data.plot.scatter('SO2','Wind')   #ชื่อตาราง.plot.scatter('x','y')
```

```
pollution_data.plot.scatter('SO2','Wind',s= 10,c= 'r') 
```

เทคนิคที่1 ใช้ขนาดของจุดแสดงปริมาณของ dimension ที่ 3 ได้

```
pollution_data.plot.scatter('SO2','Wind',s= 'RainDays') 
```

เทคนิคที่2 ใช้สีแสดงปริมาณของ dimension ที่ 3 ได้

```
pollution_data.plot.scatter('SO2','Wind',c= 'RainDays') 
```

### Scetter Matrix

```
pd.plotting.scatter_matrix(pollution_data)
```

```
pd.plotting.scatter_matrix(pollution_data); 
```

```
pd.plotting.scatter_matrix(pollution_data,alpha=1.0); #ไม่มีความโปร่งใส ทำให้ไม่รู้ว่าทับกัน
```

```
pd.plotting.scatter_matrix(pollution_data,alpha=0.1); #จุดเบาบางเกินไป สามารถปรับแอลฟาได้ขึ้นอยู่กับขนาดของดาต้า
```

```
pollution_data.plot.scatter('Temp','RainDays',s= 100,c= 'r') 
```
